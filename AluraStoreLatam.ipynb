### Importación de datos



```python
import pandas as pd

url_1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url_2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url_3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url_4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda_1 = pd.read_csv(url_1)
tienda_2 = pd.read_csv(url_2)
tienda_3 = pd.read_csv(url_3)
tienda_4 = pd.read_csv(url_4)

tienda_1.head()
```

#1. Análisis de facturación



```python
for i, tienda in enumerate([tienda_1, tienda_2, tienda_3, tienda_4], start=1):
    # Total de ingresos
    total = tienda['Precio'].sum()
    print(f'\n{"="*40}')
    print(f'Tienda {i}')
    print()
    print(f'Ingreso total: ${total:,.2f}')
```

# 2. Ventas por categoría

```python
for i, tienda in enumerate([tienda_1, tienda_2, tienda_3, tienda_4], start=1):
    total = tienda['Precio'].sum()
    print(f'\n{"="*40}')
    print(f'Tienda {i}')
    print()
    print("Productos vendidos por categoría:")
    conteo = tienda['Categoría del Producto'].value_counts()
    for categoria, cantidad in conteo.items():
      print(f'{categoria}: {cantidad}')
```

# 3. Calificación promedio de la tienda


```python
for i, tienda in enumerate([tienda_1, tienda_2, tienda_3, tienda_4], start=1):
    # Obtener el conteo de productos por categoría
    conteo = tienda['Categoría del Producto'].value_counts()
    print(f'{"="*40}')
    print(f'Tienda {i}')
    print()
    print("Calificación promedio por producto:")
    print()
    promedio_calificaciones = tienda.groupby('Categoría del Producto')['Calificación'].mean()
    for categoria, calificacion_promedio in promedio_calificaciones.items():
        print(f'{categoria}: {calificacion_promedio:.2f}')
        promedio_general = tienda['Calificación'].mean()
    print()
    print(f'Calificación promedio general de la tienda: {promedio_general:.2f}')
    print()
```

# 4. Productos más y menos vendidos

```python
for i, tienda in enumerate([tienda_1, tienda_2, tienda_3, tienda_4], start=1):
    conteo = tienda['Categoría del Producto'].value_counts()
    print(f'\n{"="*40}')
    print(f'Tienda {i}')
    print()
    print("\nProductos más y menos vendidos:")
    ventas_por_producto = tienda['Producto'].value_counts()

    producto_mas_vendido = ventas_por_producto.idxmax()
    cantidad_mas_vendida = ventas_por_producto.max()

    producto_menos_vendido = ventas_por_producto.idxmin()
    cantidad_menos_vendida = ventas_por_producto.min()

    print()
    print(f'Más vendido: {producto_mas_vendido} ({cantidad_mas_vendida} ventas)')
    print(f'Menos vendido: {producto_menos_vendido} ({cantidad_menos_vendida} ventas)')
```

# 5. Envío promedio por tienda

```python
for i, tienda in enumerate([tienda_1, tienda_2, tienda_3, tienda_4], start=1):
    conteo = tienda['Categoría del Producto'].value_counts()
    print(f'\n{"="*40}')
    print(f'Tienda {i}')
    print()
    costo_envio_promedio = tienda['Costo de envío'].mean()
    print(f'Costo de envío promedio de la tienda: ${costo_envio_promedio:.2f}')
```
