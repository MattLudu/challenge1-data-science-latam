### Importación de datos



```python
import pandas as pd

url_1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url_2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url_3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url_4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda_1 = pd.read_csv(url_1)
tienda_2 = pd.read_csv(url_2)
tienda_3 = pd.read_csv(url_3)
tienda_4 = pd.read_csv(url_4)

tienda_1.head()
```

#1. Análisis de facturación



```python
for i, tienda in enumerate([tienda_1, tienda_2, tienda_3, tienda_4], start=1):
    # Total de ingresos
    total = tienda['Precio'].sum()
    print(f'\n{"="*40}')
    print(f'Tienda {i}')
    print()
    print(f'Ingreso total: ${total:,.2f}')
```

# 2. Ventas por categoría

```python
for i, tienda in enumerate([tienda_1, tienda_2, tienda_3, tienda_4], start=1):
    total = tienda['Precio'].sum()
    print(f'\n{"="*40}')
    print(f'Tienda {i}')
    print()
    print("Productos vendidos por categoría:")
    conteo = tienda['Categoría del Producto'].value_counts()
    for categoria, cantidad in conteo.items():
      print(f'{categoria}: {cantidad}')
```

# 3. Calificación promedio de la tienda


```python
for i, tienda in enumerate([tienda_1, tienda_2, tienda_3, tienda_4], start=1):
    # Obtener el conteo de productos por categoría
    conteo = tienda['Categoría del Producto'].value_counts()
    print(f'{"="*40}')
    print(f'Tienda {i}')
    print()
    print("Calificación promedio por producto:")
    print()
    promedio_calificaciones = tienda.groupby('Categoría del Producto')['Calificación'].mean()
    for categoria, calificacion_promedio in promedio_calificaciones.items():
        print(f'{categoria}: {calificacion_promedio:.2f}')
        promedio_general = tienda['Calificación'].mean()
    print()
    print(f'Calificación promedio general de la tienda: {promedio_general:.2f}')
    print()
```

# 4. Productos más y menos vendidos

```python
for i, tienda in enumerate([tienda_1, tienda_2, tienda_3, tienda_4], start=1):
    conteo = tienda['Categoría del Producto'].value_counts()
    print(f'\n{"="*40}')
    print(f'Tienda {i}')
    print()
    print("\nProductos más y menos vendidos:")
    ventas_por_producto = tienda['Producto'].value_counts()

    producto_mas_vendido = ventas_por_producto.idxmax()
    cantidad_mas_vendida = ventas_por_producto.max()

    producto_menos_vendido = ventas_por_producto.idxmin()
    cantidad_menos_vendida = ventas_por_producto.min()

    print()
    print(f'Más vendido: {producto_mas_vendido} ({cantidad_mas_vendida} ventas)')
    print(f'Menos vendido: {producto_menos_vendido} ({cantidad_menos_vendida} ventas)')
```

# 5. Envío promedio por tienda

```python
for i, tienda in enumerate([tienda_1, tienda_2, tienda_3, tienda_4], start=1):
    conteo = tienda['Categoría del Producto'].value_counts()
    print(f'\n{"="*40}')
    print(f'Tienda {i}')
    print()
    costo_envio_promedio = tienda['Costo de envío'].mean()
    print(f'Costo de envío promedio de la tienda: ${costo_envio_promedio:.2f}')
```

# Grafico y Analisis Final

```python
import matplotlib.pyplot as plt
import pandas as pd

# Nombres y tiendas
nombres_tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
tiendas = [tienda_1, tienda_2, tienda_3, tienda_4]

# Calcular calificación promedio por tienda
calificaciones = [tienda['Calificación'].mean() for tienda in tiendas]

# Calcular ingresos totales (sumando precios)
ingresos = [tienda['Precio'].sum() for tienda in tiendas]

# Calcular costo de envío promedio por tienda
envios = [tienda['Costo de envío'].mean() for tienda in tiendas]

# GRÁFICO 1: Calificación promedio
plt.figure(figsize=(6, 4))
plt.bar(nombres_tiendas, calificaciones, color='green')
plt.title('Calificación Promedio por Tienda')
plt.ylabel('Calificación (0 a 5)')
plt.ylim(0, 5)
plt.show()

# GRÁFICO 2: Ingresos totales
plt.figure(figsize=(6, 4))
plt.bar(nombres_tiendas, ingresos, color='blue')
plt.title('Ingresos Totales por Tienda')
plt.ylabel('Ingresos en $')
plt.show()

# GRÁFICO 3: Costo de envío promedio
plt.figure(figsize=(6, 4))
plt.plot(nombres_tiendas, envios, marker='o', color='red')
plt.title('Costo de Envío Promedio')
plt.ylabel('Costo en $')
plt.grid(True)
plt.show()
```

#Análisis para  el cierre de la Tienda 4

Después de revisar los datos generales de las cuatro tiendas, se puede ver que la Tienda 4, aunque no esté muy por debajo en todos los indicadores, sí es consistentemente la que menos destaca en varios aspectos clave.

Primero, si bien vende una buena cantidad de productos y tiene una distribución de categorías parecida al resto, no lidera en ninguna. O sea, está bien equilibrada, pero no sobresale en nada. Esto la vuelve menos competitiva frente a las demás tiendas, que sí tienen puntos fuertes más definidos.

En cuanto a la facturación total, está en el último lugar. No es una diferencia abismal, pero sigue siendo la más baja. Considerando que los costos fijos seguramente son parecidos entre todas, esta diferencia puede marcar un impacto importante en la rentabilidad general del negocio.

Además, aunque tiene el costo de envío más barato, eso no está repercutiendo en un beneficio visible, como mayores ventas o una mejor experiencia para el cliente. De hecho, las calificaciones promedio de los productos están por debajo de las otras tiendas. Esto podría indicar problemas en la atención, calidad de productos o logística, lo cual es preocupante.

Por otro lado, su producto menos vendido es uno que en otras tiendas tiene mejor rendimiento, lo cual también da a pensar que no está aprovechando bien el catálogo o no llega con fuerza a su público objetivo.

Con todo esto, mantener la Tienda 4 funcionando podría estar absorbiendo recursos que podrían usarse para potenciar las otras tres tiendas, que muestran un desempeño más sólido y estable. Cerrar la Tienda 4 no se trata solo de eliminar la más débil, sino de reforzar lo que ya está funcionando mejor.
